### 특성(Feature) 컬럼 생성

```python
train['FamilySize'] = train.SibSp + train.Parch + 1
train[['SibSp', 'Parch', 'FamilySize']]
```

```python
# Embarked에 따른 컬럼을 생성
# Q, S => UK, C => Austria
# train['Nationalisty_AT'] = 
# CC = train['Embarked'] == 'C'
# SQ = train[(train['Embarked'] == 'S') | (train['Embarked'] =='Q')]
train['Nationality_AT'] = train['Embarked'].isin(['C'])
train['Nationality_UK'] = train['Embarked'].isin(['S', 'Q'])
```

```python
train1 = train.copy()
train2 = train.copy()
```

```python
train2.loc[train2['Embarked']=='C', 'Nationality'] = 'Austria'
train2.loc[train2['Embarked']!='C', 'Nationality'] = 'England'
train2.head()
```

```python
# train1['Nationality'] = 
# train1['Nationality_AT'] = True
result = []
for i in train1['Nationality_AT']:   
    result.append('Autria' if i == True else 'England')
train1['Nationality'] = result
```

```python
train1
```

```python
# Fare에 대해 분류 컬럼을 만들려고 한다. 기준점 ?
train.Fare.describe()
```

```python
print('1사분위수 :', train.Fare.quantile(.25))
print('2사분위수 :', train.Fare.quantile(.50))
print('3사분위수 :', train.Fare.quantile(.75))
print('4사분위수 :', train.Fare.quantile(1))
```

```python
# 2분위수와 3분위수의 근사 정수치를 선택
train['LowFare'] = train['Fare'] < 15
train['MediumFare'] = (train['Fare'] >= 15) & (train['Fare'] < 30)
train['HighFare'] = train['Fare'] > 30

train[['Fare', 'LowFare', 'MediumFare', 'HighFare']].head(10)
```

```python
# train2.loc[train2['Embarked']=='C', 'Nationality'] = 'Austria'
# train2.loc[train2['Embarked']!='C', 'Nationality'] = 'England'
# train2.head()

train1.loc[train['Fare'] < 15, 'FareType'] = 'Low'
train1.loc[(train['Fare'] >= 15) & (train['Fare'] < 30), 'FareType'] = 'Medium'
train1.loc[train['Fare'] > 30, 'FareType'] = 'High'
```

```python
def check_fare_type(fare):
    if fare < 15:
        return "Low"
    elif fare >= 15 and fare < 30:
        return "Medium"
    else:
        return "High"
```

```python
train['FareType2'] = train['Fare'].apply(check_fare_type)
train[["Fare", "FareType2"]].head()
```

### DataFrame Merging(병합)

```python
df1 = pd.DataFrame(
    {
        'key' : list('bbacaab'),
        'data1' : range(7)
     }
)
df1
```

```python
df2 = pd.DataFrame(
    {
        'key' : list('abd'),
        'data2' : range(3)
     }
)
df2
```

```python
# default : Inner Join
pd.merge(df1, df2, on='key')
```

```python
# 모두 보이기
pd.merge(df1, df2, on='key', how='outer')
```

```python
# df1 기준으로 합치기
pd.merge(df1, df2, on='key', how='left')
```

```python
# df2 기준으로 합치기
pd.merge(df1, df2, on='key', how='right')
```

### key name이 다를 경우

```python
df1 = pd.DataFrame(
    {
        'lkey' : list('bbacab'),
        'data1' : range(6)
    }
)
df1
```

```python
df2 = pd.DataFrame(
    {
        'rkey' : list('ababd'),
        'data2' : range(5)
    }
)
df2
```

```python
pd.merge(df1, df2, left_on='lkey', right_on='rkey', how='inner')
```

### DataFrame의 Concatenation(문자열 합치기)

```python
df1 = pd.DataFrame(
    np.arange(6).reshape(3,2),  #3행, 2열(0~5까지)
    index=['a','b','c'],
    columns=['one','two']
)
df1
```

```python
df2 = pd.DataFrame(
    5 + np.arange(4).reshape(2,2),  #3행, 2열(0~5까지)
    index=['a', 'c'],
    columns=['three','four']
)
df2
```

```python
# df1과 df2를 합치기(merge는 key값 따진다, concatenation은 안따짐)
pd.concat([df1, df2], axis=1)
```

```python
# 데이터 단순 합치기 (row로 합치기: 0, column합치기 : 1)
pd.concat([df1, df2], axis=1, ignore_index=True)
```

```python
pd.concat([df1, df2], axis=0, ignore_index=True)
```

### 데이터의 범위 정하기

```python
ages=[20,22,25,27,21,23,37,31,61,45,41,32]
bins=[18,25,35,60,100]  #범위
```

```python
pd.cut()
```

```python
cats = pd.cut(ages, bins)
cats
```

```python
# 범주 관련 Code 보기
cats.codes
```

```python
# 범주 관련 빈도수 보기
cats.value_counts()
```

```python
# category 이름 정하기
group_names = ['Youth', 'YoungAdult', 'MiddleAged', 'Senior']
cat2 = pd.cut(ages, bins, labels=group_names)
cat2.value_counts()
```
