### 2개의 y축을 갖는 그래프 작성
- 남북한 발전량
- 막대그래프는 북한의 발전량
- 선 그래프는 북한 발전량의 전년도 대비 증감률을 백분율로 표시

```python
df = pd.read_excel('../Data/남북한발전전력량.xlsx', convert_float=True) #정수를 실수로 바꾼다.
df
```

```python
# 북한 전력량만 선택
df = df.loc[5:9]
df
```

```python
# 전력량 (억㎾h) 컬럼 삭제
df.drop('전력량 (억㎾h)', axis=1, inplace=True)
df
```

```python
# 발전전력별을 index로 변경하기
df.set_index('발전 전력별', inplace=True)
df
```

```python
# 기준점을 발전 전력별로 변경
df = df.T
df
```

```python
# 합계 컬럼을 총발전량으로 변경
df = df.rename(columns={'합계' : '총발전량'})
df['총발전량-1년'] = df['총발전량'].shift(1)    # shift: 전(prev)행의 데이터를 선택, (1)개 위에 있는 것 가져옴
df
```

```python
# 증감율 컬럼 생성
df['증감율'] = (df['총발전량'] / df['총발전량-1년'] - 1) * 100
df
```

```python
plt.style.use('ggplot')

ax1 = df[['수력','화력']].plot(
    kind='bar',
    figsize=(20, 10),
    stacked=True    # 누적막대그래프
)

ax2 = ax1.twinx()

ax2.plot(
    df.index,
    df.증감율,
    ls='--',
    marker='o',
    markersize=10,
    color='green',
    label='전년대비 증감율(%)'
)

ax1.set_ylim(0,500)
ax2.set_ylim(-50,50)

ax1.set_xlabel('년도', size=10)
ax1.set_ylabel('발전량(억 kwh)')
ax2.set_ylabel('전년대비 증감율(%)')

plt.title('북한 전력 발전량(1990 ~ 2016)', size=30)
ax1.legend(loc='upper left', fontsize=20)

plt.show()
```

### 산점도
: 서로 다른 두변수의 관계를 나타낸다.

```python
plt.style.use('default')
df.plot(
    kind='scatter',
    x = 4,
    y = 0,
    c = 'coral',
    s = 10, # 산점도 점의 크기
    figsize=(10,5)
)
plt.show()
```

```python
# 실린더의 개수를 상대적 비율을 계산하여 산포도 점의 크기에 반영한다.
cylinder_size = df[1] / df[1].max() * 100   # Hyper parameter
cylinder_size
```

```python
plt.style.use('default')
df.plot(
    kind='scatter',
    x = 4,
    y = 0,
    c = 'coral',
    s = cylinder_size, # 산점도 점의 크기
    figsize=(10,5)
)
plt.show()
```

### 파이차트
- 데이터 값이 전체에 차지하는 비율을 표시

```python
df.columns = ['mpg','cylinders','displacement','horsepower','weight',
              'acceleration','model year','origin','name']
df.head()
```

```python
# 자동차의 국가별 데이터 개수를 파이차트로 표현
# 국가별 Count를 계산하기 위해
df['count'] = 1
df.head()
```

```python
# origin 열을 기준으로 그룹화, 합계
df_origin = df.groupby('origin').sum()
df_origin
```

```python
df_origin.index = ['USA', 'EU', 'JAPAN']
df_origin
```

```python
df_origin['count'].plot(
    kind='pie',
    figsize=(7, 5),      # label때문에 7:5로 사용
    autopct='%.2f%%',
    startangle=10,      # 파이차트 출력 위치 조정
    colors=['chocolate', 'bisque', 'cadetblue']
)

plt.title('Model Origin', size = 20)
plt.axis('equal')
plt.legend(
    labels=df_origin.index,
    loc='upper right'
)

plt.show()
```
