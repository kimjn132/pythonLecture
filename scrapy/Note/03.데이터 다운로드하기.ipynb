#### mov02.jpg, mov03.jpg, mov04.jpg의 이미지 여러개를 한꺼번에 가져오기

```python
import urllib.request as req

fileName = ["mov02.jpg", "mov03.jpg", "mov04.jpg"]

for i in fileName:
    
    url = "https://zeushahn.github.io/Test/images/" + i
    saveName = "../Data/" + i

    req.urlretrieve(url, saveName)
    print(i)
```

### 기상청
http://www.kma.go.kr/weather/forecast/mid-term-rss3.jsp

stnId(지역번호)
- 전국 : 108
- 서울/경기도 : 109
- 강원도 : 105
- 전라북도 : 146
- 전라남도 : 156
- 제주특별자치도 : 184

```python
# 서울/경기도 기상청 예보 데이터 가져오기 - 방법 1

import requests as req

URL = "http://www.kma.go.kr/weather/forecast/mid-term-rss3.jsp?stnId=109"
params = {'stnId':'109'}
response = req.get(URL, params=params)
print(response.status_code)    
print(response.text)
```

```python
# 서울/경기도 기상청 예보 데이터 가져오기 - 방법 2 
import urllib.request as req

url = "http://www.kma.go.kr/weather/forecast/mid-term-rss3.jsp?stnId=109"
print("url", url)

data = req.urlopen(url).read().decode("utf-8")
print(data)
```
