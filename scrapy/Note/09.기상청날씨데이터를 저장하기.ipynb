```python
import requests
from bs4 import BeautifulSoup
```

```python
req = requests.get("https://www.kma.go.kr/weather/forecast/mid-term-rss3.jsp?stnId=108")
html = req.text
soup = BeautifulSoup(html, "html.parser")
```

```python
for i in soup.find_all("location"):
    print(i.find('city'))
```

```python
#  서울로 테스트
seoul = soup.find('location')
# tmef: 날짜, wf: 날씨, tmn: 최저기온, tmx: 최고기온
for i in seoul.find_all('data'):
    print(i.find('tmef').text)
    print(i.find('wf').text)
    print(i.find('tmn').text)
    print(i.find('tmx').text)
```

```python
# 도시별 날씨 데이터 구축
weather = {}

for i in soup.find_all('location'):
    weather[i.find('city').text] = []
    for j in i.find_all("data"):
        temp = []
        temp.append(j.find('tmef').text)
        temp.append(j.find('wf').text)
        temp.append(j.find('tmn').text)
        temp.append(j.find('tmx').text)
        weather[i.find('city').text].append(temp)
```

```python
print(weather['서울'])
```

```python
print(weather['서귀포'])
```

```python
# csv file로 저장하기
# 현재 dictionary로 되어있으나 List형태로 변환이 필요
for i in weather:
    for j in weather[i]:
        print(i, j[0], j[1], j[2], j[3])
    print("-" * 20)
```

```python
# List로 생성하기
tempList = []
for i in weather:
    for j in weather[i]:
        tempList.append((i, j[0], j[1], j[2], j[3]))
tempList
```

```python
import pandas as pd
```

```python
weatherDataFrame = pd.DataFrame(tempList, columns=['지역', '날짜', '날씨', '최저온도', '최고온도'])
weatherDataFrame
```

```python
# file로 저장
weatherDataFrame.to_csv("../Data/weatherList.csv", index=False)
```

```python
import sqlite3
```

```python
areas = []
dates = []
weathers = []
lowtemps = []
hightemps = []

area = list(forecast['지역'])
date = list(forecast['날짜'])
weather = list(forecast['날씨'])
lowtemp = list(forecast['최저온도'])
hightemp = list(forecast['최고온도'])
```

```python
# for i in area:
#     areas.append(i)
for i in range(len(list(forecast['지역']))):
    areas.append(list(forecast['지역']))
    dates.append(list(forecast['날짜']))
    weathers.append(list(forecast['날씨']))
    lowtemps.append(list(forecast['최저온도']))
    hightemps.append(list(forecast['최고온도']))
```

```python
# sqlite 데이터베이스 연결하기
dbPath = "../Data/forecast.sqlite"
conn = sqlite3.connect(dbPath)    #connection

# 테이블을 생성하고 데이터 넣기
curs = conn.cursor()

# 여러 커맨드 한꺼번에 실행할 때 vs, curs.execute() - 한 커맨드씩 실행할 때
curs.executescript(
    """
    /* items 테이블이 이미 있다면 제거하기 */
    DROP TABLE IF EXISTS items;
    
    /* 테이블 생성하기 */
    CREATE TABLE items(
        id INTEGER PRIMARY KEY AUTOINCREMENT,
        area TEXT,
        date TEXT,
        weather TEXT,
        lowtemp INTEGER,
        hightemp INTEGER
    );
    """
)

# 위의 작업을 DB에 반영하기
conn.commit()
```

```python
# 데이터 넣기

curs = conn.cursor()
for i in tempList:
    curs.execute("INSERT INTO items (area, date, weather, lowtemp, hightemp) VALUES (?,?,?,?,?)", i)  #TUPLE 형식으로 insert
conn.commit()
```

```python
# 데이터 추출하기
curs = conn.cursor()
curs.execute("SELECT * FROM items")
item_list = curs.fetchall()
print(item_list)    #tuple 형식으로 들어가 있다.
```
