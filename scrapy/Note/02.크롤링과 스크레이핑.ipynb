# requests

응답코드

1xx : 요청을 받았고 작업 진행중     
2xx : 사용자의 요청이 성공적으로 수행됨     
3xx : 요청은 완료되었으나 리다이렉션(redirection)이 필요    
4xx : 사용자의 요청이 잘못됨    
5xx : 서버에 오류가 발생함.     

```python
import requests as req

URL = "https://www.naver.com"
response = req.get(URL)
print(response.status_code)     #200
```

```python
import requests as req

URL = "https://www.naver.com"
response = req.get(URL)
print(response.status_code)     #200
print(response.text)
```

```python
import requests as req

URL = "https://search.naver.com/search.naver"
params = {'query':'python'}
response = req.get(URL, params=params)
print(response.status_code)     #200
print(response.text)
```

```python
import requests as req

URL = "https://comic.naver.com/webtoon/detail.nhn"
params = {'titleId':'687915', 'no':'157'}
response = req.get(URL, params=params)
print(response.status_code)     #200
print(response.text)
```

```python
import requests as req

URL = "https://raw.githubusercontent.com/naver/naver-openapi-guide/draft/naver-openapi-swagger.json"
response = req.get(URL)
result = response.json()
print(response.status_code)
print(type(result))
print(result)
```
