```python
from bs4 import BeautifulSoup
import urllib.request as req

url = "https://zeushahn.github.io/Test/python/bs_exam01.html"

# 소스가 res에 담긴다
res = req.urlopen(url)
# print(res)

# BeautifulSoup으로 분석하기
soup = BeautifulSoup(res, "html.parser")
print(soup)
```

```python
# 원하는 부분 추출하기
#head = body 같은 레벨
h1 = soup.html.body.h1
p1 = soup.html.body.p
p2 = p1.next_sibling.next_sibling   # 두 번 하는 이유: 줄바꿈 때문에

# 요소의 글자 출력하기 : string , text
print("h1 = " + h1.text)
print("p1 = " + p1.string)
print("p2 = " + p2.string)
```

### CSS 선택자 사용하기
- BeautifulSoup은 JavaScript Library인 JQuery처럼 CSS 선택자를 지정해서 원하는 요소를 추출 가능.

```python
from bs4 import BeautifulSoup
import urllib.request as req

url = "https://zeushahn.github.io/Test/python/bs_exam04.html"

# 소스가 res에 담긴다
res = req.urlopen(url)
# print(res)

# BeautifulSoup으로 분석하기
soup = BeautifulSoup(res, "html.parser")
print(soup)
```

```python
# 필요한 부분을 CSS Query로 추출하기 (#: id, .:class, >: 자식, 빈칸(한 칸만 띄우면 됨) : 후손)
# id(unique), class(common)
h1 = soup.select_one("div#meigen > h1").string
print("h1 =", h1)

li_list = soup.select("div#meigen > ul.items > li")

for li in li_list:
    print("li=", li.string)
    
li_list = soup.select("div#meigen li")

for li in li_list:
    print("li=", li.string)
```
