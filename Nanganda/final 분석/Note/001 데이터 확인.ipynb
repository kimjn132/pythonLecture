- PCOS : 다낭성 난소 증후군(발병 여부)

- Hip : 힙 크기
- Waist : 허리 사이즈
- WHRatio : 허리와 엉덩이  비율
- Weight_gain : 체중 증가유무
- Hair_growth : 다모증 유무
- Skin_darkening : 흑색 극세포증 유무
- Hair_loss : 탈모 유무
- Pimples : 여드름 유무
- Fast_food : 패스트 푸드 섭취 유무확인
- Reg_exercise : 운동여부
- BP_systolic : 수축기  혈압
- BP_diastolic : 확장기  혈압

```python
#na값 확인하기
df_PCOS_use.isna().sum()
```

> df_PCOS_use에는 na값이 2개 있는 것을 알 수 있다.

```python
#데이터 타입을 전부 float로 바꿔주기
df_PCOS_use = df_PCOS_use.astype('float')
df_PCOS_use.info()
```

```python
# na값을 중앙값으로 채워넣기
for x in df_PCOS_use.columns[df_PCOS_use.isna().any()].tolist():
    df_PCOS_use[x] = df_PCOS_use[x].fillna(df_PCOS_use[x].median())
```

```python
#na값 다시 확인
df_PCOS_use.isna().any()
```

```python
# 칼럼 명 변경
df_PCOS_use.columns = ['PCOS', 'Age', 'Weight', 'Height', 'BMI',
       'Blood Type', 'Pulse rate', 'Cycle length',
       'Marraige Status', 'Pregnant', 'aborptions',
       'Hip', 'Waist', 'Waist:Hip Ratio', 'Weight gain',
       'hair growth', 'Skin darkening', 'Hair loss',
       'Pimples', 'Fast food', 'Reg.Exercise',
       'BP_Systolic', 'BP_Diastolic']
df_PCOS_use.head()
```

```python
#데이터 저장하기
df_PCOS_use.to_csv('../Data/pcos_colums_organize.csv', index=False)
```
