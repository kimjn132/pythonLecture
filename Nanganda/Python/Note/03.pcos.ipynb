```python
from sklearn.model_selection import train_test_split

train_input, test_input, train_target, test_target = train_test_split(x, y, test_size = 0.20, random_state = 42)
```

```python
# 정규화
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
ss.fit(train_input)

train_scaled = ss.transform(train_input)
test_scaled = ss.transform(test_input)
```

```python
# Feature를 제곱한 column을 생성하여 기존의 Column에 붙인다.
train_poly = np.column_stack((train_input ** 2, train_input))       # ** 두배 
test_poly = np.column_stack((test_input ** 2, test_input))
```

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
ss.fit(train_poly)

train_poly_scaled= ss.transform(train_poly)
test_poly_scaled = ss.transform(test_poly)
```

```python
print(train_poly_scaled.shape, test_poly_scaled.shape)
```

```python
# 로지스틱 회귀 모델
from sklearn.linear_model import LogisticRegression
# C : Cost function - 높으면 약하게 규제, 낮으면 강하게 규제
lr = LogisticRegression()
lr.fit(train_poly_scaled, train_target)

print("Train Score : ", lr.score(train_poly_scaled, train_target))
print("Test Score : ", lr.score(test_poly_scaled, test_target))
```

## 함수 만들기

```python
# bmi, Systole, Diastole 등급값 user_input에 extend(iterable)함수로 array 마지막에 하나씩(총 3개) 추가
# bmi 계산 후 해당 수치에 따라 등급 나눠주기(A, B, C)
def bmi(weight, height):
        # bmi 계산
        bmi = [(weight/((height/100)**2))]
          
        # extend로 넣어줌 
        return user_input.extend(bmi)


# 혈압 입력값에 따라 등급 숫자 추가
def Systole(p1):
      if p1<110:
        p1_c = [0]
      elif p1>=110 and p1<130:
        p1_c = [1]
      elif p1>=130  and p1<140:
        p1_c = [2]
      else:
        p1_c = [3]
        
      # extend로 넣어줌
      return user_input.extend(p1_c)

def Diastole(p2):
      if p2<60:
        p2_c = [0]
      elif p2>=60 and p2<85:
        p2_c = [1]
      elif p2>=85  and p2<90:
        p2_c = [2]
      else:
        p2_c = [3]
        
      # extend로 넣어줌 
      return user_input.extend(p2_c)
    
```
