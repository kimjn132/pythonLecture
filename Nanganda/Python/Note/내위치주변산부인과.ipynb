```python
# 가입 없이 좌표->주소 변환
from geopy.geocoders import Nominatim
```

```python
def geocoding_reverse(lat_lng_str): 
    geolocoder = Nominatim(user_agent = 'South Korea', timeout=None)
    address = geolocoder.reverse(lat_lng_str)

    return address

# 내 위치 위경도값 입력
address = geocoding_reverse('37.55586052139369, 127.1925484977747') 
print(address)
```

```python
# 구만 추출
gu = [ad for ad in address[0].split(", ") if "시" in ad]
gu
```

```python
myloc = wh[wh_address['주소'].str.contains(gu[0][0])]
myloc
```

```python
# 1. MAP 만들기
seoul_hos = folium.Map(
    # 내 위치 위경도값 입력해야함
    location=[37.5440,127.1256],
    tiles='openstreetmap', 
    zoom_start=11,
)

# 대학교 위치정보를 Marker로 표시
for add, name, tel, lat, lng in zip(myloc.주소, myloc.기관명, myloc.대표전화, myloc.병원위도, myloc.병원경도):
    
    html = """
    <style>
        .section-title {
            font-size: 36px;
            font-weight: bold;
            text-transform: uppercase;
            letter-spacing: 2px;
            color: #ff66cc;
            text-shadow: 2px 2px 4px rgba(0,0,0,0.3);
            }
        .section-content {
            background: linear-gradient(to bottom, #fff, #f1f1f1);
            box-shadow: 2px 2px 5px rgba(0, 0, 0, 0.3);
            padding: 15px;
            position: relative;
            }
        .section-content p {
            font-size: 13px;
            line-height: 1.5;
            }
        @media (max-width: 768px) {
        .section-content {
                padding: 5px;
            }
        }
    </style>
    <div class="section-title">
        <h5>%s</h5>
    </div>
    <div class="section-content">
        <p>
            <b>주소:</b> %s<br>
            <b>전화번호:</b> %s<br>
        </p>
    </div>
    """% (name, add, tel)
    
    
    folium.Marker([lat, lng],
                popup=folium.Popup(html, max_width=300),
                icon=folium.Icon(color='blue', icon='medkit', prefix='fa'),
                title=name,
                ).add_to(seoul_hos)

    # folium.Marker([lat, lng],
    #               popup=f"{name}\n{add}",
    #               popup_orientation = 'vertical',
                  
    #               # prefix='fa'는 Font Awesome library, glyphicon: Glyphicon library
    #               icon=folium.Icon(color='blue', icon='medkit', prefix='fa')
    #               ).add_to(seoul_hos)

# 기존의 마커와 모양이 다르다. CircleMarker는 범위를 주어서 사용할 수 있는 마커이다.
# 대학교 위치정보를 CircleMarker로 표시
for name, lat, lng in zip(myloc.기관명, myloc.병원위도, myloc.병원경도):
    folium.CircleMarker([lat, lng],
                        radius=5, # radius 는 원의 반지름 크기이다.
                        color = 'brown', # 원 둘레의 색상
                        fill=False, # 칼라를 채우겠냐 채우지 않겠냐. (원형안의 색상)
                        # fill_color='coral', # 원형마커의 색상(원을 채우는 색)
                        # fill_opacity = 0.7, # 원형의 투명도 1은 완전한 불투명이다.
                        popup=name,
                        ).add_to(seoul_hos)

seoul_hos.save("../Data/seoul_hos.html")

seoul_hos
```
