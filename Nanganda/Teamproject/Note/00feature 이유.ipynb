```python
# replace = False(중복 없이)
df_normal_sample = df[df['PCOS']==0.0].sample(n=177, replace=False)
```

```python
df_normal_sample
```

```python
df_abnormal = df[df['PCOS']==1.0]
```

```python
df2 = pd.concat([df_normal_sample, df_abnormal])
```

```python
177+177
```

```python
df2.reset_index(inplace=True)
```

```python
df2
```

```python
df2['PCOS'].value_counts().plot(
    kind='bar',
    rot=0,
    color = ['blue','red']
)
plt.xlabel("PCOS (Y/N)")
plt.ylabel("Frequency")
plt.show()
```

```python
fig = plt.figure(figsize=(20,20))
ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
ax4 = fig.add_subplot(2,2,4)

# 선형회귀선 표시
sns.regplot(
    x ='PCOS',
    y='Age',
    data=df2,
    scatter_kws={'color':'black'}, # scatter color
    line_kws={'color':'red'}, #regression line color
    ax=ax1,
    fit_reg=True # 선형 회귀선을 보이게 함. 기본값이 True. 즉 False를 안써주면 보임

)

sns.regplot(
    x ='PCOS',
    y='Weight',
    data=df2,
    scatter_kws={'color':'black'}, # scatter color
    line_kws={'color':'red'}, #regression line color
    ax=ax2,
    fit_reg=True # 선형 회귀선을 보이게 함. 기본값이 True. 즉 False를 안써주면 보임

)

sns.regplot(
    x ='PCOS',
    y='Hip',
    data=df2,
    scatter_kws={'color':'black'}, # scatter color
    line_kws={'color':'red'}, #regression line color
    ax=ax3,
    fit_reg=True # 선형 회귀선을 보이게 함. 기본값이 True. 즉 False를 안써주면 보임

)

sns.regplot(
    x ='PCOS',
    y='Waist',
    data=df2,
    scatter_kws={'color':'black'}, # scatter color
    line_kws={'color':'red'}, #regression line color
    ax=ax4,
    fit_reg=True # 선형 회귀선을 보이게 함. 기본값이 True. 즉 False를 안써주면 보임

)

plt.show()

#선에 퍼짐이 있는것은 데이터가 적어서 오차가 있을 수 있다는 의미
```
