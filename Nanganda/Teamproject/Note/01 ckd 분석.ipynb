# 원핫인코딩

### 숫자인데 object로 되어있는 데이터를 numeric으로 변경

```python
ckd['pcv'] = pd.to_numeric(ckd['pcv'], errors='coerce')
ckd['wc'] = pd.to_numeric(ckd['wc'], errors='coerce')
ckd['rc'] = pd.to_numeric(ckd['rc'], errors='coerce')
```

### 데이터 타입이 object인 칼럼명 추출

```python
object_cols =  [cols for cols in ckd.columns if ckd[cols].dtype == 'object']
object_cols
```

### 위에서 뽑은 각 칼럼들의 데이터 값 보여줌

```python
for cols in object_cols:
    print(ckd[cols].sort_values(ascending=0).unique())
```

### 뽑은 각 데이터 값들을 1 or 0 으로 변환

```python
ckd['rbc'].replace(to_replace = {'normal': 1, 'abnormal': 0},inplace=True)
ckd['pc'].replace(to_replace = {'normal': 1, 'abnormal': 0},inplace=True)
ckd['pcc'].replace(to_replace = {'present': 1, 'notpresent': 0},inplace=True)
ckd['ba'].replace(to_replace = {'present': 1, 'notpresent': 0},inplace=True)
ckd['htn'].replace(to_replace = {'yes': 1, 'no': 0},inplace=True)
ckd['dm'].replace(to_replace = {'yes': 1, 'no': 0},inplace=True)
ckd['cad'].replace(to_replace = {'yes': 1, 'no': 0},inplace=True)
ckd['appet'].replace(to_replace = {'poor': 1, 'good': 0},inplace=True)
ckd['pe'].replace(to_replace = {'yes': 1, 'no': 0},inplace=True)
ckd['ane'].replace(to_replace = {'yes': 1, 'no': 0},inplace=True)
ckd['classification'].replace(to_replace = {'ckd': 1, 'notckd': 0},inplace=True)



ckd
```

# 결정트리

```python
ckd_data = ckd.drop(columns='classification')
ckd_data.head(3)
```

### Feature 과 Target 나눠주기

```python
# Feature, Target
data = ckd_data.to_numpy()
target = ckd['classification'].to_numpy()
```

### Train 데이터와 Test 데이터 나눠주기

```python
# Train, Test
from sklearn.model_selection import train_test_split
train_input, test_input, train_target, test_target = \
    train_test_split(data, target, test_size=0.2, random_state=42)
```

### 결정트리 돌려보기

```python
from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier(random_state=42)
dt.fit(train_input, train_target)
print('Train score :', dt.score(train_input, train_target))
print('Test score :', dt.score(test_input, test_target))
```

### 가지치기(Tree Pruning)
- max_depth의 parameter의 값을 조절한다.
- decision tree의 과대적합을 막는다.

```python
dt = DecisionTreeClassifier(max_depth=1, random_state=42)
dt.fit(train_input, train_target)

print('Train score :', dt.score(train_input, train_target))
print('Test score :', dt.score(test_input, test_target))
```

> Train score가 너무 높고 과대적합이므로 다른 모델을 사용해본다

# RandomForest

```python
# RandomForest

from sklearn.model_selection import cross_validate
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_jobs=-1, random_state=42)
scores = cross_validate(rf, train_input, train_target, return_train_score=True, n_jobs=-1)
print(np.mean(scores['train_score']))
print(np.mean(scores['test_score']))
```

```python
# 부트스트랩 결정시 남는 샘플(oob:out of back)도 특성으로 구분할 수 있다.
rf = RandomForestClassifier(oob_score=True, n_jobs=-1, random_state=42)
rf.fit(train_input, train_target)
rf.oob_score_
```

> oob를 사용해도 100%의 예측이 나온다  
> 따로 검증셋을 구성하지 않아도 oob로 검증세트의 역할을 대신할 수 있다.

### Extra Tree

```python
from sklearn.ensemble import ExtraTreesClassifier

et = ExtraTreesClassifier(n_jobs=-1, random_state=42)
scores = cross_validate(et, train_input, train_target, return_train_score=True, n_jobs=-1)
print(np.mean(scores['train_score']), np.mean(scores['test_score']))
```

### Gradient Boosting(그레디언트 부스팅)

```python
from sklearn.ensemble import GradientBoostingClassifier
gb = GradientBoostingClassifier(random_state=42)
scores = cross_validate(gb, train_input, train_target, return_train_score=True, n_jobs=-1)
print(np.mean(scores['train_score']), np.mean(scores['test_score']))
```

```python
gb = GradientBoostingClassifier(n_estimators=500, learning_rate=0.2, random_state=42)
scores = cross_validate(gb, train_input, train_target, return_train_score=True, n_jobs=-1)
print(np.mean(scores['train_score']), np.mean(scores['test_score']))
```

```python
gb.fit(train_input, train_target)
gb.feature_importances_
```

```python
from sklearn.inspection import permutation_importance

gb.fit(train_input, train_target)
result = permutation_importance(gb, train_input, train_target, n_repeats=10, random_state=42, n_jobs=-1)
result.importances_mean
```
