# 차원확대

```python
dfpcos = df.copy()
```

```python
def BMI_Insights(x):
    
      if x<18:
        return 'A'
      elif x>=18 and x<26:
        return 'B'
      else:
        return 'C'
dfpcos['BMI_catg']=dfpcos['BMI'].map(BMI_Insights)


def Systole(p1):
      if p1<110:
        return 'hypo'
      elif p1>=110 and p1<130:
        return 'normal'
      elif p1>=130  and p1<140:
        return 'pre_hyper'
      else:
        return 'hyper'

dfpcos['Systolic_cond']=dfpcos['BP _Systolic (mmHg)'].map(Systole)


def Diastole(p2):
      if p2<60:
        return 'hypo'
      elif p2>=60 and p2<85:
        return 'normal'
      elif p2>=85  and p2<90:
        return 'pre_hyper'
      else:
        return 'hyper'

dfpcos['Diastolic_cond']=dfpcos['BP _Diastolic (mmHg)'].map(Diastole)
```

```python
# x값 만들어주기
x=df_back.iloc[:,1:]
x.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split

train_input, test_input, train_target, test_target = train_test_split(x, y, test_size = 0.20, random_state = 42)
```

```python
# 정규화
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
ss.fit(train_input)

train_scaled = ss.transform(train_input)
test_scaled = ss.transform(test_input)
```

```python
# 로지스틱 회귀 모델
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(train_input, train_target)

print("Train Score : ", lr.score(train_input, train_target))
print("Test Score : ", lr.score(test_input, test_target))
```

```python
lr.fit(train_scaled, train_target)
print("Train Score : ", lr.score(train_scaled, train_target))
print("Test Score : ", lr.score(test_scaled, test_target))
```

### SVM

```python
from sklearn import svm
clf = svm.SVC()
clf.fit(train_input, train_target)
print('train score: ',clf.score(train_input, train_target))
print('test score:', clf.score(test_input, test_target))
```

```python
clf = svm.SVC()
clf.fit(train_scaled, train_target)
print('train score: ',clf.score(train_scaled, train_target))
print('test score:', clf.score(test_scaled, test_target))
```
