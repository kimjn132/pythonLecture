```python
import pandas as pd
```

```python
csv = pd.read_csv("../Data/iris.csv")
```

```python
csv_data = csv[['SepalLength', 'SepalWidth', 'PetalLength', 'PetalWidth']]
csv_label = csv['Name']
```

```python
from sklearn.model_selection import train_test_split
train_data, test_data, train_label, test_label = train_test_split(csv_data, csv_label, stratify=csv_label)
```

### RandomForest

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier()      # instance
clf.fit(train_data, train_label)
clf.score(test_data, test_label)
```

# 독버섯과 관련된 데이터를 사용한 머신러닝

```python
# 데이터 다운 받기
import urllib.request as req
local = "../Data/mushroom.csv"
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/mushroom/agaricus-lepiota.data'
req.urlretrieve(url, local)
print('OK')
```

```python
mr = pd.read_csv("../Data/mushroom.csv", header=None)
mr.head()
```

```python
mr[2].unique()
```

```python
# 문자를 숫자(ASCII Code)로 변경 / chr()
ord('x')
chr(120)
```

```python
#  데이터 내부의 기호를 숫자로 변환하기
label = []
data = []

for row_index, row in mr.iterrows():
    label.append(row.loc[0])
    row_data = []
    for v in row.loc[1:]:
        row_data.append(ord(v))
    data.append(row_data)
    
print(data[0:20])
```

# label을 DataFrame으로 만들기
label

```python
#label을 DataFrame으로 만들기
labelTemp = pd.DataFrame(label)
labelTemp.head()
```

```python
dataTemp = pd.DataFrame(data)
dataTemp.rename(columns=lambda x: x+1, inplace=True)
dataTemp.head()
```

```python
# 2개의 DataFrame 합치기
mr2 = pd.concat([labelTemp, dataTemp], axis='columns')
mr2.head()
```

```python
csv_data = dataTemp
csv_label = labelTemp
```

```python
from sklearn.model_selection import train_test_split
train_data, test_data, train_label, test_label = train_test_split(csv_data, csv_label, stratify=csv_label)
```

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier()      # instance
clf.fit(train_data, train_label)
clf.score(test_data, test_label)
```

```python
# 답
train_data, test_data, train_label, test_label = train_test_split(mr2.loc[:,1:], mr2.loc[:,0],stratify=mr2.loc[:,0])

clf = RandomForestClassifier()      # instance
clf.fit(train_data, train_label)
clf.score(test_data, test_label)
```

```python
# 적용
for i in range(1, len(mr2.columns)):
    dataTemp = pd.get_dummies(data=dataTemp, columns=[i], prefix=str(i))        # prefix = 문자(컬럼이름)
    
dataTemp.head()
```

```python
# 데이터 합치기 
mr3 = pd.concat([labelTemp, dataTemp], axis='columns')
mr3.head()
```

```python
train_data, test_data, train_label, test_label = train_test_split(mr3.iloc[:,1:], mr3.iloc[:,0],stratify=mr3.iloc[:,0])

clf = RandomForestClassifier()      # instance
clf.fit(train_data, train_label)
clf.score(test_data, test_label)
```
