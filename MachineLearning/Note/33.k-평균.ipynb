### k-Means
- 대표적인 군집 알고리즘으로 군집의 중심을 찾는 작업
- 클러스터(cluster) : 군집
- 클러스터링 : 군집화 하는 작업
- 과일 데이터에는 k가 3인 작업으로 k를 hyper parameter
- 평균 데이터와 비슷하다
- 같은 데이터는 똑같은(하나의) 데이터로 취급한다.

```python
fruits = np.load("../Data/fruits.npy")
fruits.shape
```

```python
# 3차원 데이터를 2차원 데이터로 변경하기 
fruits_2d = fruits.reshape(-1, 100*100)
fruits_2d.shape
```

```python
# k-Means Model
from sklearn.cluster import KMeans
```

```python
km = KMeans(n_clusters=3, random_state=42)
km.fit(fruits_2d)
```

```python
# 군집화한 Label 출력
km.labels_
```

```python
km.labels_
```

```python
# 군집화한 Cluster 개수와 확인
np.unique(km.labels_, return_counts=True)   # return_counts : unique한 개수를 보여 준다.
```

```python
# Cluster 0이 111개인지 확인
count0 = 0
for i in km.labels_:
    if i==0:
        count0+=1
        
print(count0)       #unique에서 0번이 111개라는 뜻 
```

```python
def draw_fruits(arr, ratio=1):  #ratio: 해상도 크기
    n = len(arr)    # n : 샘플 개수
    # 한줄에 10개씩 이미지를 그립니다. 샘플 개수를 10으로 나누어 전체 행 개수를 계산
    rows = int(np.ceil(n/10))   # 올림
    # 행이 1개이며 열 개수는 샘플 개수이고 그렇지 않으면 10개 입니다.
    cols = n if rows < 2 else 10
    
    fig, axs = plt.subplots(rows, cols, figsize=(cols*ratio, rows*ratio), squeeze=False)
    
    for i in range(rows):
        for j in range(cols):
            if i*10+j < n:
                axs[i,j].imshow(arr[i*10+j], cmap='gray_r')
            axs[i,j].axis('off')
plt.show()
```

```python
draw_fruits(fruits[km.labels_ == 0])
```

```python
draw_fruits(fruits[km.labels_ == 1])
```

```python
draw_fruits(fruits[km.labels_ == 2])
```

### 클러스터 중심

```python
km.cluster_centers_.reshape(-1, 100, 100)
```

```python
draw_fruits(km.cluster_centers_.reshape(-1, 100, 100), ratio=3)
```

```python
# 100번째의 클러스터 중심 찾기
km.transform(fruits_2d[100:101])
```

```python
# 100번째의 예측 군집값
km.predict(fruits_2d[100:101])
```

```python
draw_fruits(fruits[100:101])
```

```python
# 현 k-means의 반복횟수
km.n_iter_      # 숫자가 적으면 데이터가 좋은 거다.
```

---
### 최적의 k값 찾기

```python
scores = []
for k in range(2, 10):
    km = KMeans(n_clusters=k, random_state=42)
    km.fit(fruits_2d)
    scores.append(km.inertia_)

plt.plot(range(2,10), scores)
plt.show()
```

> 제일 처음 꺽이는 부분이 제일 좋은 값이다 : 여기서는 3

Elbow Method : Cluster간의 거리의 합을 나타내는 inertia(관성)가 급격히 떨어지는 구간이 생기는데     
이 지점의 k값을 군집의 개수로 사용.
